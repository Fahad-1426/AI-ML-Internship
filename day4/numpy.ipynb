```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print(arr)
```

```python

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Addition
print("Addition:", arr1 + arr2)

# Multiplication
print("Multiplication:", arr1 * arr2)
```

```python
arr = np.arange(1, 10)

new_arr = arr.reshape(3, 3)

print(new_arr)
```

```python
arr = np.array([10, 20, 5, 8])

print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))
```
